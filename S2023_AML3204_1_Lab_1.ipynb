```python
from googleapiclient.discovery import build
import requests
from IPython.display import Image, display
```

```python
# Youtube API Key
API_KEY = ''
```

```python
def get_video_comments(vidId, apikey):
  count = 0
  # Create YT resource object
  my_yt = build('youtube', 'v3', developerKey=API_KEY)

  # Retrieve YT video comments
  video_responses = my_yt.commentThreads().list(
                                              part='snippet, replies',
                                              videoId=vidId).execute()


  # Iterate over the comments
  for item in video_responses['items']:
    count += 1
    print("Comment-",count,':')
    comment = item['snippet']['topLevelComment']['snippet']['textDisplay']
    author = item['snippet']['topLevelComment']['snippet']['authorDisplayName']
    authorImage = item['snippet']['topLevelComment']['snippet']['authorProfileImageUrl']
    display(Image(requests.get(authorImage).content))
    print(author, ':', comment, '\n')
```

```python
vid = 'ujQFbBeMoYA'
get_video_comments(vid, API_KEY)
```
