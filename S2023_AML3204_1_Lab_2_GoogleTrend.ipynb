```python
import pandas as pd
from pytrends.request import TrendReq
```

```python
pytrnd = TrendReq(retries=5)
```

```python
pytrnd.build_payload(kw_list=['BTS']) #Lionel Messi'])
df = pytrnd.interest_by_region()
df
```

```python
df_indexed = df.reset_index()
df_indexed
```

```python
df_sorted = df_indexed.sort_values('BTS', ascending=False) #Lionel Messi
df_sorted
```

```python
df_sorted.head(20)
```

## Generate df from Car data

```python
brands = ['BMW','Mercedes','Ford','Toyota','Honda']

# Build payload with parameters
pytrnd.build_payload(kw_list=brands,
                    timeframe='today 12-m',
                    geo='CA')
```

```python
df_ca = pytrnd.interest_by_region()
df_ca
```

```python
df_ca.plot(figsize=(12,6), y=brands)
```

```python
df_ca.plot(figsize=(12,6), y=brands, kind='bar')
```
